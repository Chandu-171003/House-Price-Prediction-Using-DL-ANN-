```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('/content/Housing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.hist(df['price'], bins=30)
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.title("Distribution of House Prices")
plt.show()
```

```python
num_cols = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']
plt.figure(figsize=(4,2))
sns.boxplot(data=df[num_cols])
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(4,2))
df['furnishingstatus'].value_counts().plot(kind='bar')
plt.xlabel("Furnishing Status")
plt.ylabel("Count")
plt.title("Furnishing Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(4,2))
df['parking'].value_counts().sort_index().plot(kind='barh')
plt.xlabel("Count")
plt.ylabel("Parking Spaces")
plt.title("Parking Availability")
plt.show()
```

```python
plt.figure(figsize=(5,3))
df['mainroad'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.ylabel("")
plt.title("Houses Connected to Main Road")
plt.show()
```

```python
plt.figure(figsize=(5,3))
df['airconditioning'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.ylabel("")
plt.title("Air Conditioning Availability")
plt.show()
```

```python
df_sorted = df.sort_values('area')
plt.figure(figsize=(4,2))
plt.plot(df_sorted['area'], df_sorted['price'])
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.title("Price Trend with Area")
plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.scatter(df['area'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs House Price")
plt.show()
```

```python
plt.figure(figsize=(4,2))
avg_price = df.groupby('bedrooms')['price'].mean()
avg_price.plot(kind='bar')
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.title("Average Price by Bedrooms")
plt.show()
```
